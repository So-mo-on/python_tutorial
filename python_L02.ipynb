#function

```python
# Function
def function(x):  # Defining a function
    y = x + 2
    z = y ** 2
    return z  # Result of the function

print(function(3))  # Returns (3+2)^2
```

#Important libraries!

There are many libraries in Python for different scientific purposes, such as numpy, scipy, pandas, networkx, scikit-learn, and more.\
These libraries provide various tools. One of the most useful libraries is numpy, which is perfect for high-dimensional computing.

```python
import numpy as np  # Importing the library. You can use the name np (or any alias) instead of numpy.

x = [1, 2, 3, 4]  # This is a Python list
y = np.array(x)  # Convert Python list to numpy array. Numpy array is similar to vectors in linear algebra.

print(function(y))  # The function will be applied element-wise on the numpy array

print('shape of y =', np.shape(y))  # Print the shape of the array y
z = np.array([[1, 2], [3, 4]])  # Create a 2x2 numpy array
print(z)
print('shape of z =', np.shape(z))  # Print the shape of the array z

# You can also reshape y to z
z = np.reshape(y, (2, 2))
print('shape of z =', np.shape(z))  # Print the shape of the reshaped array z

print(2 * z)  # Element-wise multiplication
print(2 + z)  # Element-wise addition
```

#Plots

Matplotlib is a great library for drawing all kinds of plots.

```python
import matplotlib.pyplot as plt

x = np.arange(0, 10, 0.1)  # Generate numbers in the range between 0 to 10 with step size = 0.1
y = np.sin(x)  # Compute the sine of each element in x

plt.plot(x, y, label='sin(x)')  # Plot y = sin(x)
plt.xlabel('x')  # Label for x-axis
plt.ylabel('sin(x)')  # Label for y-axis
plt.title('Title: sin(x) graph')  # Title of the plot
plt.legend()  # Show legend
plt.show()  # Display the plot
```

#Solution

```python
# Write a function to calculate the factorial of numbers.
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

print(factorial(5))  # Output: 120
```

```python
# Write a function is_prime that takes a number as an argument and returns True if the number is prime and False otherwise.
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

print(is_prime(7))  # Output: True
print(is_prime(10))  # Output: False
```

```python
import numpy as np

# Create a NumPy array with random numbers and find the mean, median, standard deviation, and variance.
random_numbers = np.random.rand(10)

mean_value = np.mean(random_numbers)
median_value = np.median(random_numbers)
std_deviation = np.std(random_numbers)
variance_value = np.var(random_numbers)

print(f"Random Numbers: {random_numbers}")
print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Standard Deviation: {std_deviation}")
print(f"Variance: {variance_value}")
```

```python
# Write a function normalize that takes a NumPy array as an argument and returns a normalized version of the array (values scaled to be between 0 and 1).
def normalize(arr):
    arr_min = np.min(arr)
    arr_max = np.max(arr)
    return (arr - arr_min) / (arr_max - arr_min)

# Example usage
array = np.array([1, 2, 3, 4, 5])
normalized_array = normalize(array)
print(normalized_array)  # Output: [0.  0.25 0.5  0.75 1. ]
```

```python
# Write a function basic_stats that takes a NumPy array and returns a dictionary with the mean, median, standard deviation, and variance.
def basic_stats(arr):
    stats = {
        'mean': np.mean(arr),
        'median': np.median(arr),
        'standard_deviation': np.std(arr),
        'variance': np.var(arr)
    }
    return stats

# Example usage
array = np.random.rand(10)
stats = basic_stats(array)
print(stats)
```
